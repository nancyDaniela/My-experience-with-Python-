```python
# Manejo de datatypes
from pyspark.sql.types import * 

#Hace la conexion entre el codigo y el spark server
from pyspark.sql import SparkSession

#metodos de Spark y los asigna a la variable F
import pyspark.sql.functions as F
import missingno as msno

### Main
spark = SparkSession.builder.appName("primer_codigo").getOrCreate()

# transferir archivo desde carpeta archivo separado por comas csv
archivo="datasets/super_hero_info.csv"

df = spark.read.csv( archivo, header=True, sep=",", inferSchema=True)
#df.show()


## Filtrar  los superheroes de Marvel y los guardan en un dataframe de nombre df_marvel
df_marvel = df.filter ((df.Publisher=="Marvel Comics"))

## Filtrar los superheroes de DC Comics y los guardan en un dataframe de nombre df_dc
df_dc = df.filter ((df.Publisher=="DC Comics"))

## Del dataframe df_marvel filtrar quienes son good ( y guardarlo en df_marvel_buenos) y filtrar son bad ( y guardarlo en df_marvel_malos)
df_marvel_buenos= df_marvel.filter ((df_marvel.Alignment=="good"))
df_marvel_malos= df_dc.filter ((df_dc.Alignment=="bad"))

## Del dataframe de df_dc filtrar quienes son Male ( y guardarlo en df_dc_men) y filtrar son Female ( y guardalo en df_dc_women)
df_dc_men= df_dc.filter ((df_dc.Gender=="Male"))
df_dc_women= df_dc.filter ((df_dc.Gender=="Female"))

## Del dataframe de df_dc_women, quienes son Human y tambi√©n Mutant
df_dc_women.filter((df_dc_women.Race=="Human") & (df_dc_women.Race=="Mutant")).show()

## Del dataframe df_marvel_buenos, quienes no tienen asignado Eye color o Hair color o Skin color
df_marvel_buenos.filter((df_marvel_buenos["Eye Color"]=="-") | (df_marvel_buenos["Hair Color"]=="-") | (df_marvel_buenos["Skin Color"]=="-")).show()

## Del dataframe df_dc_women, quien tiene Weight  y Heigth con valor negativo
df_dc_women.filter((df_dc_women["Weight"]<0)&(df_dc_women["Height"]<0)).show()
```
