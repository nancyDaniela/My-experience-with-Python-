```python
# VERIFICACION DE DATOS NULOS

import pandas as pd
import missingno as msno

archivo="/content/drive/MyDrive/datasets/iris.csv"
df=pd.read_csv(archivo, sep=",", header=0) 

# Por codigo
df.isnull().sum()

# Por missingno
df.isnull().sum()
ax=msno.bar(df)
ax.get_figure().show()


# VERIFICACION DE DATOS NO VALIDOS PARA COLUMNAS NUMERICAS
for i in df.columns:
  print(i," --> ",pd.to_numeric(df[i], errors="coerce").notnull().all())


# NO SE PRESENTO ALGUN PROBLEMA DE DATOS NULOS, DEBIDO A QUE 
# NO SE ENCONTRO UN DATO NULO O ALGUN CARACTER RARO
```
