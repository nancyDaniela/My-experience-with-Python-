Del archvio de catalogo ocupar la columna Magnitud para agregar una nuva columna para clasificar el registro en:


*   0 a 1.9: como no sentido
*   2.0 a 3.9: como débil
*   4.0. a 4.9 Moderado
*   5.0 a 5.9: como fuerte
*   6.0 a 6.9:: como violento
*   7.0 en adelante: como severo

En base a la fecha, agregar una columna nueva para asignar una etiqueta nueva a aquellos resitros que se presenten en cualquier día del mes, y cualquier años, pero etiquetanto con el NOMBRE DEL MES QUE SE PRESENTA (NO NUMERO).

También ocupando  la fecha, agregar una columna nueva para asignar una etiqueta nueva a  aquellos resitros que se presenten en 19 de   del mes de septiembre de cualquier año, etiquetarlos en la nueva columna con el valor de 19S


Guardando todos las modificaciones en un archvio separado por comas

Tener en cuenta que se debe validar lso datos nulos y valores mal asignados en columnas numericas.

```python
import pandas as pd

archivo="/content/drive/MyDrive/datasets/catalogo.csv"
df=pd.read_csv(archivo, sep=",", header=0) 


#0 a 1.9: como no sentido
#2.0 a 3.9: como débil
#4.0. a 4.9 Moderado
#5.0 a 5.9: como fuerte
#6.0 a 6.9 como violento
#7.0 en adelante: como severo

#for i in df.columns:
#  print(i,"->",pd.to_numeric(df[i], errors="coerce").notnull().all())

#df.Magnitud.unique()
df=df.replace({"Magnitud":{"no calculable":0}}) 
df["Magnitud"] = df["Magnitud"].astype('float64')

rango=[0,1.9,3.9,4.9,5.9,6.9,9.9]
etiqueta=["No sentido","Debil", "Moderado", "Fuerte", "Violento","Severo"]
df["clasificacion"]=pd.cut(df.Magnitud, bins=rango, labels=etiqueta)
df


#En base a la fecha, agregar una columna nueva para asignar una etiqueta nueva 
#a aquellos resitros que se presenten en cualquier día del mes, y cualquier años, pero etiquetanto con el NOMBRE 
#DEL MES QUE SE PRESENTA (NO NUMERO).

df["Mes"] = (pd.to_datetime(df["Fecha"], format='%Y.%m.%d', errors="coerce").dt.month_name(locale="en_US.UTF-8"))
df

#También ocupando la fecha, agregar una columna nueva para asignar una etiqueta nueva a 
#aquellos resitros que se presenten en 19 de del mes de septiembre de cualquier año, etiquetarlos en la nueva columna con el valor de 19S

df["Fecha"] = df["Fecha"].astype("datetime64")

df["Dia"]=df["Fecha"].apply(lambda r:r.day)
sep=df["Mes"].apply(lambda x:"September" in x)

dia=df["Dia"].isin([19])
df["19S"]=dia & sep
#a=df["19S"]==True
#df[a].count() #total:958 
df


#Verificacion de datos nulos y rellenado 
df["clasificacion"] = df["clasificacion"].astype("object")
#df.isnull().sum() #hay 17649 nulos en la col clasificacion
re_na={"clasificacion":"No clasificado aun"}
df=df.fillna(re_na)
df.isnull().sum()

df
#Guardando todos las modificaciones en un archvio separado por comas
#df.to_csv("df_SSN.csv")

```
